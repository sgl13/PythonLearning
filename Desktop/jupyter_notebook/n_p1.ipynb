```python
#check the version of the version
import numpy as np
print(np.__version__)
```

```python
l = [1,2,3,4,5,6]
#create an array from above the list
a = np.array(l)
print(a)
#find the shape of the array
print(a.shape)
#find the dimension of the array
print(a.ndim)
#find the number of elements in the array
print(a.size)
#find the element data types from above the array
print(a.dtype)
#find the each element size in bytes
print(a.itemsize)
```

```python
#create 2 arrays with the same aize apply addition operation 2 arrays
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)
#when we add 2 array's of same size is known as vector addition
```

```python
a = np.array([1,2,3])
b = np.array([4])
print(a+b)
"""broadcasting method is a powerful mechanism where single element of an array adds with multiple or 
differant size of another array"""
```

```python
#dot product / inner product - vector multiplication of 2 arrays and sum of the result
#what is the rule to apply dot product on 1D array?
#the arrays must be in same size
a = np.array([1,2])
b = np.array([3,4])
d = np.dot(a,b)
print(d)
```

```python
c = a@b
print(c)
```
