The **print()** function is used to output text, numbers, or variable values to the console.

```python
# Print to the console
print("hello world")
```

```python
# Print to the console
print(22)
```

**Simple Arithmetic**:Python can be used as a calculator to perform standard mathematical operations.

```python
# Simple arithmetic
(5+4)/3
```

**Variables** are containers used to store data values in memory.

```python
# Assign variables
country = "india"
age = 30
```

```python
print(country)
print(age)
```

**Comparison operators** compare two values and return a **Boolean** (`True` or `False`):

```python
# Evaluations
# Double equals signs are used to check equivalency.
10*3 == 40
```

**Conditional statements** execute different blocks of code based on whether a condition is `True` or `False`.

```python
# Conditional statements
if age>=18:
    print('adult')
else:
    print('minor')
```

**Loops** are used in Python to **repeat a block of code** multiple times automatically without writing the same code over and over again.

A `for` loop is used to iterate over a sequence (such as a list of numbers or items) and execute a block of code for each item.

```python
# Loops
for number in [1,2,3,4,5]:
    print(number)
```

```python
# Loops
my_list = [3, 6, 9]

for x in my_list:
    print(x / 3)
```

A **function** is a block of organized, reusable code that is used to perform a single, related action. Functions help make your code modular, readable, and easy to debug.

```python
# Functions
def is_adult(age):

    if age >= 18:
        print('adult')
    else:
        print('minor')
```

```python
# Use the function that was just created.
is_adult(14)
```

```python
# Use the built-in sorted() function.
new_list = [20, 25, 10, 5]

sorted(new_list)
```

**Lists** store multiple items in a single variable. They are ordered and changeable.

**append(item)**: Adds a new item to the end of the list.
**len(list)**: Returns the total number of elements.
**Iteration**: Use a `for` loop to step through items one by one.

```python
# Create a list of skills
skills = ["Python", "SQL", "Git"]

# Add a new skill
skills.append("Power BI")

# Check total number of skills
print("Total skills:", len(skills))

# Loop through the list to print each skill
for skill in skills:
    print("Skill:", skill)
```

**Dictionaries** store data in `{key: value}` pairs, ideal for structured records.

**Access value**: `dict["key_name"]`
**Add/Update**: `dict["new_key"] = value`
**Iterate through pairs**: `for key, value in dict.items():`

```python
# Create a dictionary to store person information
person = {
    "name": "Shivu",
    "role": "Developer",
    "location": "Karnataka"
}

# Accessing values using keys
print("Name:", person["name"])
print("Role:", person["role"])

# Adding a new key-value pair
person["language"] = "Python"

# Loop through key-value pairs
for key, value in person.items():
    print(f"{key.capitalize()}: {value}")
```

**Tuples** (Immutable Collections)
Tuples store multiple items in a single variable using parentheses `()`. Unlike lists, tuples are **immutable** (their elements cannot be changed or modified after creation).
**Syntax:** `my_tuple = (1, 2, 3)`
**Use case:** Storing data that should not be altered (e.g., coordinates, configurations,aadharcard).

```python
coordinates = (12.9716, 77.5946)
print("Latitude:", coordinates[0])
print("Longitude:", coordinates[1])
```

**Sets** (Unordered & Unique Collections)
Sets store multiple items using curly braces `{}`. They are **unordered** and **automatically remove duplicate values**.
* **Syntax:** `my_set = {1, 2, 3}`
* **Methods:** `.add()` to insert elements, `.remove()` to delete elements.
**Use case:** Removing duplicates from data and checking membership efficiently.

```python
skills_set = {"Python", "SQL", "Git", "Python"}
skills_set.add("Power BI")

print("Unique Skills Set:", skills_set)
print("Is SQL in skills?", "SQL" in skills_set)
```
