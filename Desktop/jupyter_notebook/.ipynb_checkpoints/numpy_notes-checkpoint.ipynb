#the central object in numpy array's

```python
l = [1,2,3,4,5,6]
#create an array from above the list
a = np.array(l)
print(a)
#find the shape of the array
print(a.shape)
#find the dimension of the array
print(a.ndim)
#find the number of elements in the array
print(a.size)
#find the element data types from above the array
print(a.dtype)
#find the each element size in bytes
print(a.itemsize)
```

#elements of math operations

```python
#create 2 arrays with the same aize apply addition operation 2 arrays
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)
#when we add 2 array's of same size is known as vector addition
```

```python
a = np.array([1,2,3])
b = np.array([4])
print(a+b)
"""broadcasting method is a powerful mechanism where single element of an array adds with multiple or 
differant size of another array"""
```

```python
#dot product / inner product - vector multiplication of 2 arrays and sum of the result
#what is the rule to apply dot product on 1D array?
#the arrays must be in same size
a = np.array([1,2])
b = np.array([3,4])
d = np.dot(a,b)
print(d)
```

```python
c = a@b
print(c)
```

#applying condition on array #04-05-2026
#boolean indexing(true or false)

```python
import numpy as np
```

```python
a = np.array([10,21,16,15,17,18,20])
#check the elements in the above array even or odd
print(a%2==0)
```

```python
a = np.array([10,21,16,15,17,18,20])
print(a>20)
```

```python
l = [10,21,16,15,17,18,20]
#check the element in the above array even or odd
for i in l:
  if i%2==0:
    print("even")
  else:
    print("odd")
  
```

#apply condition in array's
#np.where(logic,true,false)

```python
a = np.array([10,21,16,15,17,18,20])
print(np.where(a%2==0,a,-1))
```

```python
#print the indexes where the condition is true
#np.argwhere(logic)
print(np.argwhere(a%2==0))
```

```python
#reshaping of an array
#np.reshape(rows,columns) converts the data into 2d array
a = np.array([1,2,3,4,5,6])
#convert the above array into 2 columns and 3 rows
print(a.reshape(3,2))
```

```python
#concatination of array's
#np.concat((arr1,arr2),axis=0/1))
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])
print(np.concat((a,b),axis=1))
```

```python
#Hstac:- adds sequence of array horizentally
print(np.hstack((a,b)))
```

```python
#Vstack:- adds sequence of array vertically
print(np.vstack((a,b)))
```

```python
#copying of an array's
a = np.array([1,2,3])
b = a.copy()
b[1]=6
print(a)
print(b)
#np.copy() :- if we make changes in the new array it won't impact on old array
```

```python
#generating the array's
```

```python
#generate an array of zeros with 2row and 3 columns
#syntax:- np.zeros((rows,column))
a = np.zeros((2,3))
print(a)
```

```python
#syntax:- np.ones((rows,column))
#generate an array of ones with 2row and 3 columns
b = np.ones((2,3))
print(b)
```

```python
#generate an array of specific number (6) with 2row and 3 columns
#syntax:- np.full((rows,col),number)
c = np.full((2,3),6)
print(c)
```

```python
#generate an array of identity matrix 
#syntax:- np.eye()
d = np.eye(5)
print(d)
```

```python
#split an a range of array in to "n" equal parts
#syntax:- np.linspace(start,end,partitions)
e = np.linspace(0,10,5)
print(e)
```

```python
# generate an array of random numbers that is uniformly distributed
a = np.random.random((2,3))
print(a)
```

```python
#generate an array that is normally distributed
#q) when we suppose to say a data is normally distributed?
#when the mean(avg) is equal to zero and standard deviation is eual to 1
b = np.random.randn(10,5)
print(b)
```

```python
b = np.random.randn(10,5)
print(b.mean(),b.std())
```

```python
#generate an random integer with 3rows and 3columns
c = np.random.randint(1,10,size=(3,3))
print(c)
```
